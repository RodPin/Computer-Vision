```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,text='HELLO',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255),thickness=26)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
res = cv2.erode(img,kernel,iterations=4)
display_img(res)
```
