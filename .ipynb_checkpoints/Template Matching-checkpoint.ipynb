```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('./imgs/daddy.jpeg')
full = cv2.cvtColor(full,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
crop = full[30:200,330:450]
plt.imshow(crop)
```

```python
methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_SQDIFF']
```

```python
method1= eval(methods[2])
res= cv2.matchTemplate(full,crop,method1)
```

```python
plt.imshow(res)
```

```python
for m in methods:
    fullCopy=full.copy()
    
    method= eval(m)
    
    res = cv2.matchTemplate(fullCopy,crop,method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    
    height,width,channels= crop.shape
    
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(fullCopy,top_left,bottom_right,(255,0,0),10)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HeatMap')
      
        
```
