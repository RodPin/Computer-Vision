```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('./imgs/parede.jpg')
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
##load_img()
```

```python
img = load_img()
display_img(img)
```

```python
##quanto maior o gama mais escuro fica
gamma = 1/4
img_gamma = np.power(img/255,gamma)
display_img(img_gamma)
```

```python
###Displaying a text on the image
img=load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='Hello',org=(100,600),fontFace=font,fontScale=10,color=(2,233,23),thickness=4)
display_img(img)
```
