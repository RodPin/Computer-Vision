```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('./flower.jpg')
img2 = cv2.imread('./surf.jpg')
```

```python
img1 =  cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
[img1.shape,
img2.shape]
```

```python
##turning the 2 images to the same size
resized_img1= cv2.resize(img1,(800,800))
resized_img2= cv2.resize(img2,(800,800))
##printing it
[resized_img1.shape,
resized_img2.shape]
```

```python
blended = cv2.addWeighted(src1=resized_img1,alpha=0.5,src2=resized_img2,beta=0.5,gamma=0)
plt.imshow(blended)
                          
```

```python
############################################################
################      OVERLAYS    ##########################
############################################################
```

```python
img_maior=img1
img_menor=img2
x_offset=60
y_offset=80
##shape[1] = x
##shape[0] = y
## .shape  = Y X COR
x_end = x_offset + img_menor.shape[1]
y_end = y_offset + img_menor.shape[0]
```

```python
## overlaying (sobrepondo) os pixels
img_maior[y_offset:y_end,x_offset:x_end] = img_menor
plt.imshow(img_maior)
```
